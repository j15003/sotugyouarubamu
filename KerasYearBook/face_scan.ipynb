```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
%matplotlib inline
```

```python
#入力ファイルのパスを指定
in_jpg = "./OpenCV_test/test02.jpg"
out_jpg = "./OpenCV_test/test_1.jpg"
```

```python
#入力画像の表示
plt.show(plt.imshow(np.asarray(Image.open(in_jpg))))
```

```python
# 画像の読み込み 
image_gs = cv2.imread(in_jpg) 
```

```python
# グレースケールに変換
#image_gs = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# 顔認識用特徴量ファイルを読み込む --- （カスケードファイルのパスを指定）
cascade = cv2.CascadeClassifier("opencv-master/data/haarcascades/haarcascade_frontalface_alt.xml")
```

```python
# 顔認識の実行
face_list = cascade.detectMultiScale(image_gs,scaleFactor=1.1,minNeighbors=1,minSize=(100,100))
```

```python
#顔が１つ以上検出された時
if len(face_list) > 0:
    for rect in face_list:
        image_gs = image_gs[rect[1]:rect[1]+rect[3],rect[0]:rect[0]+rect[2]]
        #顔が検出されなかった時
else:
    print("no face")
```

```python
cv2.imwrite(out_jpg, image_gs)
```

```python
#出力画像の表示
plt.show(plt.imshow(np.asarray(Image.open(out_jpg))))
```

### 顔検出[画像複数版]

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys, os
from PIL import Image
%matplotlib inline
```

```python
#入力ファイルのパスを指定
in_jpg = "./picture/"
out_jpg = "./out_picture/"
```

```python
#リストで結果を返す関数
def get_file(dir_path):
    filenames = os.listdir(dir_path)
    return filenames
```

```python
pic = get_file(in_jpg)
```

```python
print(pic) 
```

```python
for i in pic:
    #写真名のファイルの作成
    #out_path = out_jpg + i
    #os.mkdir(out_path)
    # 画像の読み込み 
    image_gs = cv2.imread(in_jpg + i)
 
    # 顔認識用特徴量ファイルを読み込む --- （カスケードファイルのパスを指定）
    #cascade = cv2.CascadeClassifier("opencv-master/data/haarcascades/haarcascade_frontalface_alt.xml")
    cascade = cv2.CascadeClassifier("opencv-master/data/haarcascade_frontalface_default.xml")
    
    
    # 顔認識の実行
    face_list = cascade.detectMultiScale(image_gs,scaleFactor=1.1,minNeighbors=1,minSize=(1,1))
    
    # 顔だけ切り出して保存
    no = 1;
    for rect in face_list:
        x = rect[0]
        y = rect[1]
        width = rect[2]
        height = rect[3]
        dst = image_gs[y:y + height, x:x + width]
        #save_path = out_path + '/' + 'out_('  + str(i) +')' + str(no) + '.jpg'
        save_path = out_jpg + 'out_('  + str(i) +')' + str(no) + '.jpg'
        
        print(x,y,width,height)
        
        #認識結果の保存
        a = cv2.imwrite(save_path, dst)
        plt.show(plt.imshow(np.asarray(Image.open(save_path))))
        print(no)
        no += 1
```
